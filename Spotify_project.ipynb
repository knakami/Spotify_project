```python
# Check data types of each column.
df_tracks.dtypes  
# Oh no, release_date is an object ! we should change it to datetime.
```

```python
# Convert the 'release_date' column to datetime format.
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce')

# Create a new column with the formatted release_date in the YYYY-MM-DD format.
df_tracks['formatted_date'] = df_tracks['release_date'].dt.strftime('%Y-%m-%d')

# Set the 'formatted_date' column as the index.
df_tracks.set_index('formatted_date', inplace=True)

# Display the resulting DataFrame with the new index.
df_tracks.head()
```

```python
# Delete the 'release_date' column because it is no longer needed.
df_tracks.drop('release_date', axis=1, inplace=True)
```

```python
# Display information about a specific artist.
df_tracks[['artists']].iloc[3]   # returns a Pandas Series (with only 2 columns, one is an index)
```

```python
# Convert units in the 'duration' column from milliseconds to seconds.
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x:round(x/1000))

# Remove column 'duration_ms' because it is no longer needed.
df_tracks.drop(['duration_ms'], inplace=True, axis=1)  
df_tracks.head()
```

```python
# Before creating the heatmap, I need to create a new dataframe in which only numeric values are present (non-numeric values will be removed).
corr_df = df_tracks.drop(['key', 'mode', 'explicit', 'id', 'name', 'artists', 'id_artists', 'time_signature'], axis = 1).corr(method = 'pearson')
corr_df.head(12)   # it shows correlation matrix with values from -1 to 1
```

```python
# Create a heatmap in seaborn that displays the correlation between all numeric values.
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='magma', linewidths=1, linecolor='white')
plt.title('Correlation Heatmap Between Variables')
plt.show()
```

```python
# Create a random sample, it returns a new sample after running this code. Check with sample_df.head()
sample_df = df_tracks.sample(int(0.004 * len(df_tracks)))
sample_df.head()
```

# <font size="4">Correlation:</font>
- Correlation measures the strength and direction of a linear relationship between two quantitative variables.
- Correlation is expressed by the correlation coefficient, which can take values from -1 to 1.
- A value of 1 indicates a perfect positive correlation, -1 a perfect negative correlation, and 0 no linear correlation.

# <font size="4">Regression:</font>
- Regression focuses on predicting the value of one variable (dependent) based on the values of another variable (independent).
- The goal is to find the optimal regression line that best describes the relationship between variables.
- Regression analysis can be linear or nonlinear and can be used for modeling and prediction.

```python
# Positive correlation between the loudness of songs and their energy.
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='m', line_kws={'color': 'orange'}).set(title='Loudness vs Energy')
plt.show()
```

```python
# Positive correlation between the loudness of songs and their popularity.
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y='loudness', x='popularity', color='m', line_kws={'color': 'orange'}).set(title='Loudness vs Popularity')
plt.show()
```

```python
# Positive correlation between the loudness of songs and their danceability.
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y='loudness', x='danceability', color='m', line_kws={'color': 'orange'}).set(title='Loudness vs Danceability')
plt.show()
```

```python
# Complicated relationship between popularity and acousticness. With increasing popularity, it is likely 
# that the acousticness value decreases, and vice versa. The correlation coefficient expresses a linear 
# relationship but can be influenced by other factors and may not be fully linear.
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='m', line_kws={'color': 'orange'}).set(title='Popularity vs Acousticness')
plt.show()
```

```python
# Extract only values of years from the 'formatted_date' column and create a new column 'years'.
df_tracks['dates'] = df_tracks.index.get_level_values('formatted_date')
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
df_tracks['years'] = df_tracks['dates'].dt.year
df_tracks.head()
```

```python
# Create a histogram that displays the count of songs according to individual years.
# In 2020, the music industry was truly productive.
plt.figure(figsize=(10, 6))
sns.histplot(data=df_tracks, x='years', discrete=True, color='orange')  # axis y is implicitely set on 'count'
plt.title('Number of Songs per Year')   
plt.show()
```
